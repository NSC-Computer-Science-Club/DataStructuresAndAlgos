# Binary Search Algorithm

## What problem is it trying to solve?

- How can we **efficiently** find a value in a collection of values?

## A little bit about preconditions:

The binary search algorithm only works when the data in the collection is
already sorted. These type of requirements are normally referred as **preconditions**,
they should be documented in your code so that other developers using it
are aware of it.

In Java you would normally put such things inside your doc-comments (/\*\* comment \*/).
The method's doc comments in Python are written inside a triple quotation
string, right after your function declaration (see below implementation).

## Enough talking Jardi! Just tell me about binary search:

If you have ever used a dictionary, the dead tree version, you have used a
binary search algorithm without even knowing it.

Seriously... Go grab a printed dictionary (do they still exist?) and look for the
word **Computer**. My guess is that you won't start from the beginning, sequentially
searching until you find it.

You would normally open a page where you think that word could be; let's say
that's the middle. You find yourself in the middle of the letter **L**. You know
that **Computer** should be before that page, so you go a look into the lower
half of the dictionary but this time you open a page in the letter **F**, you
look again in the lower half but you go passed it and now you are somewhere
in the letter **B**, you look in the upper half of that sub-section. You keep
doing this until you finally find the word you are looking for.

That is the basis for the binary search algorithm, if you know in which half of
the collection your target value should be, you can discard the other half. You
are essentially cutting the data set in half every time you look. Here is a brief
description of the algorithm follow by a python implementation:

Given a sorted array and a target value:
- While you haven't exhausted all your options:
1. Look at the element in the middle of the array.
2. If your target value should be in the lower half look in the lower half.
3. If your target value should be in the upper halp look in the upper half.
4. Else, you are looking at the target value: Return it's index position.
- Return -1 (since this is not a valid index) if the value is not found.

The following is a Python implementation of the binary search algorithm:

```python
import matplotlib.pyplot as plt
import numpy as np
import sys

def binarySearch(arr, target):
    """ Search for the target value in the array and return the index position.
        Return -1 if target is not found.

        pre: The data in the array must be sorted.
        post: The data in the array is unchanged.

        return: The positional index of the element in the array, if it is found
                in the array. Else, return -1.
    """
    left = 0
    right = len(arr) - 1

    possiblePosition = 0
    while (left <= right):
        middle = (right + left) // 2
        if (target < arr[middle]):
            right = middle - 1
        elif (target > arr[middle]):
            left = middle + 1
        else:
            return middle

    return -1

arr = [2, 4, 5, 8, 10, 11, 14, 18, 20, 21, 22, 30, 40]

print("---------------- Testing binary search function ----------------")
print("Array:", arr)
print("Search for 8:", binarySearch(arr, 8))
print("Search for 2:", binarySearch(arr, 2))
print("Search for 40:", binarySearch(arr, 40))
print("Search for 9:", binarySearch(arr, 9))
print("Search for 100:", binarySearch(arr, 100))
print("Search for 0:", binarySearch(arr, 0))
print("Search for 19:", binarySearch(arr, 19))
```

# A little bit about time complexity

We talked about time complexity, and we said that the time complexity of the
binary search algorith is O(log n) (more precisely logarith base 2 of n). The
reason for this is that every time you look at an element you can cut the problem
in half by discarding the other half where you are sure not to find your target.

Here is a table with roughly (+1 in some cases) exemplifies the number of elements
you have look at before you can either find an element, or determine that it's
not there depending on the size of the data set:

| Size in 2^n   | Size of data set  | Max value checks  |
| ------------- | ----------------  | ----------------  |
| 2^1           | 2                 | 1                 |
| 2^2           | 4                 | 2                 |
| 2^3           | 8                 | 3                 |
| 2^4           | 16                | 4                 |
| ...           | ...               | ...               |
| 2^10          | 1024              | 10                |
| 2^11          | 2048              | 11                |
| 2^12          | 4096              | 12                |

Notice how as the data doubles, the number of value checks increases only by 1.

The following graphs are to give you an example on the growth rate for 3 different
time complexities as the data set grows larger:
- O(Log n) (base 2)
- O(n)
- O(n^2)

Play with this, create a graph for O(n * log n) for example.

```python
x = np.arange(2, 4096, 2)

plt.plot(x, np.log2(x), 'r')
plt.title("O(Log n): Logarithmic time complexity")
plt.show()

plt.plot(x, x)
plt.title("O(n): Linear time complexity")
plt.show()

plt.plot(x, x**2, 'b')
plt.title("O(n^2): Quadratic time complexity")
plt.show()
```
